## Question 5:

```python
import matplotlib.pyplot as plt
```

```python
first_ten_rows = mlmdata.head(10)
```

Import Matplotlib to create the line plot. Then create a variable to display the first 10 rows of the dataset.

```python
ratings = first_ten_rows['rating']
```

The ratings variable takes movie ratings from 'rating' and gets the 1st ten rows by using the first_ten_rows variable.

```python
plt.figure(figsize=(8,6))
```

```python
plt.plot(range(1, 11), ratings, marker='o', linestyle='-')
```

```python
plt.title('First Ten Movie Ratings')
```

```python
plt.xlabel('Observation')
```

```python
plt.ylabel('Rating')
```

```python
plt.xticks(range(1, 11))
```

```python
plt.grid(True)
```

```python
plt.show()
```

Creates a line plot for the first 10 ratings. 

I don't see much of a pattern with this data, mostly because of the data size. It is not wise to look for a pattern with only 10 ratings. The sample size seems to be small to conclude what the most popular rating is. Using more ratings can be better suitable to make observations. From this data, there are 6 five star ratings, 3 four star ratings, and 1 three star rating. The line plot looks jagged with drops and rises between most points. 

## Question 6:

```python
mlmdata_first_100 = mlmdata.head(100)
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.scatter(mlmdata_first_100['rating'], mlmdata_first_100['age'], alpha=0.5)
```

```python
plt.title('Ratings vs Age Scatter Plot')
```

```python
plt.xlabel('Rating')
```

```python
plt.ylabel('Age')
```

```python
plt.grid(True)
```

```python
plt.show()
```

Above creates a scatterplot for rating vs age of the first 100 rows.

Based on the scatter plot, age does not seem to affect ratings given. For example, for the 3-5 ratings, most of the ages seem to give the same rating. There are some outliers with four ratings below or at a rating of 2. Overall, age does not impact the ratings given, since the plot shows similar ratings given regardless of age.
